```python
print(len(dmerge))
def group(data,column):
    gp = data.groupby(column)    
    result=[gp.get_group(x) for x in gp.groups]
    return result
new=group(dmerge,'YES Transition')
print(len(new[0]),len(new[1]),len(new[2]))
print(new[0]['YES Transition'])
no_info=new[0]

#only use following
non_par=new[1]
par=new[2]
```

2.helper function: count and to_dataframe_

```python
def dcount(data,column):
    count=data[column].value_counts()
#print(code_count)
    title=count.index.tolist()
    value=count.tolist()
    return title,value
def to_dataframe(data1,data2,column):
    t1,v1=dcount(data1,column)
    t2,v2=dcount(data2,column)
    df1=pd.DataFrame({column:t1,"data1":v1})
    df=pd.DataFrame([v1],columns=t1)
    df2=pd.DataFrame({column:t2,"data2":v2})
    merged=pd.merge(df1,df2,how='outer',on=column)
    title=list(merged[column])
    d1=list(merged['data1'])
    d2=list(merged['data2'])
    final=pd.DataFrame([d1,d2],columns=title)
    return final
```

***Participant vs Non-participant***

```python
def to_d(d11,d12,d21,d22,column):
    t11,v11=dcount(d11,column)
    t12,v12=dcount(d12,column)
    df11=pd.DataFrame({column:t11,"data11":v11})
    df12=pd.DataFrame({column:t12,"data12":v12})
    
    t21,v21=dcount(d21,column)
    t22,v22=dcount(d22,column)
    df21=pd.DataFrame({column:t21,"data21":v21})
    df22=pd.DataFrame({column:t22,"data22":v22})
    
    merged1=pd.merge(df11,df12,how='outer',on=column)
    merged2=pd.merge(df21,df22,how='outer',on=column)
    
    total=pd.merge(merged1,merged2,how='outer',on=column)
    title=list(total[column])
    d11=list(total['data11'])
    d12=list(total['data12'])
    d21=list(total['data21'])
    d22=list(total['data22'])

    final=pd.DataFrame([d11,d12,d21,d22],columns=title)
    return final
dfinal=to_d(no_less,less,pno_less,pless,off_type)

dfin=dfinal.set_index([pd.Series(['Non-par: No','Non-par: Yes', 'Par: No','Par: Yes'])])
plot6=dfin.plot.bar(figsize=(15,7),title="MSO Offense Type vs. Committed less than 6 months")
```
