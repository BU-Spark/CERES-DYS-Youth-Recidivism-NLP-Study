```python
def check_x_y(col_x,col_y,final_col):
    conditions = [
    ((test_df[col_x] != 'missing') &(test_df[col_y] != 'missing') & (test_df[col_x]==test_df[col_y])),
 ((test_df[col_x] != 'missing') &(test_df[col_y] != 'missing') & (test_df[col_x]!=test_df[col_y])),
 ((test_df[col_x] != 'missing') &(test_df[col_y] == 'missing')),((test_df[col_x] == 'missing') &(test_df[col_y] != 'missing'))]
    choices = [test_df[col_x],'**information not equal**' ,test_df[col_x],test_df[col_y]]
    test_df[final_col] = np.select(conditions, choices, default='**no information anywhere**')
    if '**information not equal**' in test_df[final_col].values:
            return str(final_col)
    else:
            return 0
    #print (test_df[final_col].value_counts())
```

```python
df2 = test_df.columns.get_values()
col_name_list=df2.tolist()
wrong_list=[]
for col_name in col_name_list:
    if col_name.endswith('_x'):
        main_name=col_name[:-2]
        col_name_y=str(main_name)+"_y"
        final_name=str(main_name)
        #print ("Check "+str(main_name)+" Now:")
        name=check_x_y(col_name,col_name_y,final_name)
        if name !=0:
            wrong_list.append(name)
        #flag_not_equal(final_name)
print ("Here is the Abnormal List:")
print (wrong_list)
```
